```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import scatter
from sklearn.decomposition import PCA
```

```python
# 生成数据
X=np.random.randn(1000,3)
print(X.shape)
```

```python
# 使用PCA进行降维，将3维降为2维
pca=PCA(n_components=2)#"2"对应2维
X_pca=pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 可视化
# PCA转换前的3维可视化
fig=plt.figure(figsize=(12,4))
ax1=fig.add_subplot(121,#1行2列第1个子图
                    projection='3d'#指定为3D坐标
                    )
ax1.scatter(X[:,0],X[:,1],X[:,2],
            c="g" # 颜色绿色
            )
ax1.set_title("Before PCA(3D)")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")
ax1.set_zlabel("Feature 3")

# PCA转换后的2维可视化
ax2=fig.add_subplot(122) # 默认是2D
ax2.scatter(X_pca[:,0],X_pca[:,1],
            c="g" # 颜色绿色
            )
ax2.set_title("After PCA(2D)")
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")
```

```python
# 手动构建线性相关三组特征数据
n=1000
# 定义主成分方向向量
pc1=np.random.normal(0,1,n)
pc2=np.random.normal(0,0.2,n)
#不重要的第三主成分（噪声）
noise=np.random.normal(0,0.05,n)
# 构建三个特征
X=np.vstack((pc1+pc2,pc1-pc2,pc2+noise)).T
print(X.shape)
```

```python
# 使用PCA进行降维
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 可视化
# 转换前的3维可视化
fig=plt.figure(figsize=(12,4))
ax1=fig.add_subplot(121,#1行2列第1个子图
                    projection='3d'#指定为3D坐标
                    )
ax1.scatter(X[:,0],X[:,1],X[:,2],c="g")
ax1.set_title("Before PCA(3D)")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")
ax1.set_zlabel("Feature 3")
#转换后的2维可视化
ax2=fig.add_subplot(122)
ax2.scatter(X_pca[:,0],X_pca[:,1],c='g')
ax2.set_title("After PCA(2D)")
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")
plt.show()
```
