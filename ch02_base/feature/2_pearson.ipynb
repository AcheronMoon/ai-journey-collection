```python
import pandas as pd
```

```python
# 读取数据
advertising = pd.read_csv('../data/advertising.csv')
print(advertising.head())
print(advertising.describe())
print(advertising.shape)
```

```python
# 数据预处理
# 去掉第一列ID
advertising.drop(advertising.columns[0], axis=1, inplace=True)  #axis=1:"对每行进行"，inplace=True:"原地操作"
# 去掉空值
advertising.dropna(inplace=True)
# 提取特征和标签（目标值）
X = advertising.drop("Sales", axis=1)
y = advertising["Sales"]
print(X.shape, y.shape)
```

```python
# 计算皮尔逊相关系数
print(X.corrwith(y, method="pearson"))  # corrwith
```

```python
corr_matrix = advertising.corr(method="pearson")
print(corr_matrix)  # corr相关系数矩阵
```

```python
# 将相关系数矩阵画成热力图
import seaborn as sns
import matplotlib.pyplot as plt

#热力图
sns.heatmap(corr_matrix,  #相关系数矩阵
            annot=True,  #"显示数字"
            cmap="coolwarm",  #"颜色显示为冷热图"
            fmt=".2f")  #"保留两位小数"
plt.title('Feature Correlation Matrix')
plt.show()
```
