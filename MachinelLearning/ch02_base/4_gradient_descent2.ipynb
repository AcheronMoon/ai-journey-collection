```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#定义目标函数:设fx=x2，求x为何值时，fx=2
def g(x):
    return (x ** 2 - 2) ** 2


#定义梯度函数（导函数)
def gradient(x):
    return 4 * x ** 3 - 8 * x
```

```python
# 用列表保存点的变化量
x_list = []
y_list = []
```

```python
#定义超参数和x的初始值
# alpha = 0.1  # 48次
alpha = 0.01  # 147次
x = 1
```

```python
while np.abs(grad := gradient(x)) > 1e-10:
    y = g(x)
    #记录当前点的坐标
    x_list.append(x)
    y_list.append(y)
    print(f"x={x}，g(x) = {y}")
    #计算梯度
    # grad = gradient(x)
    #跟新参数
    x = x - alpha * grad
print(len(x_list))
```

```python
#画图
x = np.arange(0.9, 1.6, 0.01)
plt.plot(x, g(x))
plt.plot(x_list, y_list, "r")
plt.scatter(x_list, y_list, color="red")
plt.show()
```

```python
# 局部放大对比
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
x1 = np.arange(0.9, 1.6, 0.01)
ax[0].plot(x1, g(x1))
ax[0].plot(x_list, y_list, "r")
ax[0].scatter(x_list, y_list, color="red")
x2 = np.arange(1.40, 1.42, 0.001)
ax[1].plot(x2, g(x2))
ax[1].plot(x_list[1:], y_list[1:], "r")
ax[1].scatter(x_list[1:], y_list[1:], color="red")
plt.show()
```
