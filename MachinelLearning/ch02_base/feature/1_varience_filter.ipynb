```python
import numpy as np
```

```python
# 构造特征
a = np.random.randn(100)
print(np.var(a))  # .var(参数)返回方差
```

```python
# b = np.random.randn(100) * 0.1
b=np.random.normal(5,0.1,size=100)   #.normal(均值,标准差,size=样本数量)
print(np.var(b))
```

```python
# 构造特征向量（输入X）
X=np.vstack((a,b)).T
print(X)
print(X.shape)
```

```python
# 低方差过滤
from sklearn.feature_selection import VarianceThreshold # 导入库
vt=VarianceThreshold(0.01) # 创建过滤方差小于0.01的VarianceThreshold的对象
X_filtered=vt.fit_transform(X)
print(X_filtered)
print(X_filtered.shape)
```
