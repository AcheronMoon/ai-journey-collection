```python
import pandas as pd

# 定义数据
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]
```

```python
X = pd.DataFrame(X)  #转数据格式
y = pd.DataFrame(y)
print(X.shape, y.shape)
```

```python
# 计算斯皮尔曼相关系数
print(X.corrwith(y, method='spearman'))
```
