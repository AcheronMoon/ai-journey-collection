```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = [[2, 1], [3, 1], [1, 4], [2, 6]]
```

```python
#定义归一化类的对象
scaler=MinMaxScaler(feature_range=(-1,1))
#应用缩放器
X_scaler=scaler.fit_transform(X)
print(X_scaler)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
print(X)
```

```python
#标准化(API)
scaler=StandardScaler()
X_scaler=scaler.fit_transform(X)
print(X_scaler)
```

```python
#标准化计算验证
import numpy as np
X=np.array(X)
print(X)
mean=np.mean(X,axis=0)
std=np.std(X,axis=0)
print("均值：",mean)
print("标准差：",std)
print("标准化结果:\n",(X-mean)/std)
```
