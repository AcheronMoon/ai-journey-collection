```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
#准备数据
X = [[2, 1], [3, 1], [1, 4], [2, 6]]  # 特征
y = [0.5, 0.33, 4,3]  # 标签
```

```python
#定义模型
knn = KNeighborsRegressor(n_neighbors=2,weights='distance')
knn.fit(X, y)
#预测
x=[[4,9]]
x_pred = knn.predict(x)
print(x_pred)
```
